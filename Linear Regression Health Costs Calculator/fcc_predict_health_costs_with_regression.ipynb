```
# Load and examine the data
dataset = pd.read_csv('insurance.csv')
print(dataset.head())
print(dataset.info())
```

```
# Convert categorical data to numbers
dataset['sex'] = dataset['sex'].map({'female': 0, 'male': 1})
dataset['smoker'] = dataset['smoker'].map({'no': 0, 'yes': 1})

# Use pd.get_dummies() for region and join with the original dataset
region_dummies = pd.get_dummies(dataset['region'], prefix='region')
dataset = pd.concat([dataset, region_dummies], axis=1)
dataset.drop('region', axis=1, inplace=True)
```

```
# Split the data into train and test sets
train_dataset = dataset.sample(frac=0.8, random_state=0)
test_dataset = dataset.drop(train_dataset.index)

# Separate features from labels
train_labels = train_dataset.pop('expenses')
test_labels = test_dataset.pop('expenses')

# Normalize numerical data
scaler = StandardScaler()
train_dataset = pd.DataFrame(scaler.fit_transform(train_dataset), columns=train_dataset.columns)
test_dataset = pd.DataFrame(scaler.transform(test_dataset), columns=test_dataset.columns)

# Convert to float32 for TensorFlow
train_dataset = train_dataset.astype('float32')
test_dataset = test_dataset.astype('float32')
train_labels = train_labels.astype('float32')
test_labels = test_labels.astype('float32')

print("\nProcessed train_dataset:")
print(train_dataset.head())
print("\nProcessed train_labels:")
print(train_labels.head())
```

```
def build_model():
    # Create a Sequential model
    model = keras.Sequential([
        # Add a Dense layer with 64 units, ReLU activation, and input shape based on the training dataset
        layers.Dense(64, activation='relu', input_shape=[len(train_dataset.keys())]),
        # Add another Dense layer with 64 units and ReLU activation
        layers.Dense(64, activation='relu'),
        # Add a final Dense layer with a single unit (output layer)
        layers.Dense(1)
    ])

    # Define the optimizer with a learning rate of 0.001
    optimizer = tf.keras.optimizers.Adam(0.001)

    # Compile the model with mean squared error loss, the defined optimizer,
    # and evaluation metrics of mean absolute error and mean squared error
    model.compile(loss='mse',
                  optimizer=optimizer,
                  metrics=['mae', 'mse'])

    # Return the compiled model
    return model
```

```
# Display training progress by printing a single dot for each completed epoch
class PrintDot(keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs):
        # Print a new line every 100 epochs for better readability
        if epoch % 100 == 0:
            print('')
        # Print a dot for each completed epoch
        print('.', end='')
```

```
# Build the model using the previously defined build_model function
model = build_model()

# Set the number of epochs for training
EPOCHS = 1000

# Define an early stopping callback to stop training when the validation loss has stopped improving
early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)

# Train the model using the training dataset and labels
# Split the training data into 80% for training and 20% for validation
# Set verbose to 0 to prevent printing output for each epoch
# Use early stopping and PrintDot callback to monitor training progress
history = model.fit(
    train_dataset, train_labels,
    epochs=EPOCHS, validation_split=0.2, verbose=0,
    callbacks=[early_stop, PrintDot()])

# Visualize the model's training progress by converting the history.history dictionary to a DataFrame
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
print('\n')
print(hist.tail())  # Print the last few entries of the training history DataFrame
```

```
# Plot the training and validation Mean Absolute Error (MAE) over epochs
plt.figure(figsize=(12, 6))

# First subplot: MAE over epochs
plt.subplot(1, 2, 1)
plt.xlabel('Epoch')
plt.ylabel('Mean Abs Error [expenses]')
plt.plot(hist['epoch'], hist['mae'], label='Train Error')
plt.plot(hist['epoch'], hist['val_mae'], label='Val Error')
plt.legend()

# Second subplot: Mean Squared Error (MSE) over epochs
plt.subplot(1, 2, 2)
plt.xlabel('Epoch')
plt.ylabel('Mean Square Error [$expenses^2$]')
plt.plot(hist['epoch'], hist['mse'], label='Train Error')
plt.plot(hist['epoch'], hist['val_mse'], label='Val Error')
plt.legend()

# Display the plots
plt.show()
```

```
# RUN THIS CELL TO TEST YOUR MODEL. DO NOT MODIFY CONTENTS.
# Test model by checking how well the model generalizes using the test set.
loss, mae, mse = model.evaluate(test_dataset, test_labels, verbose=2)

print("Testing set Mean Abs Error: {:5.2f} expenses".format(mae))

if mae < 3500:
  print("You passed the challenge. Great job!")
else:
  print("The Mean Abs Error must be less than 3500. Keep trying.")

# Plot predictions.
test_predictions = model.predict(test_dataset).flatten()

a = plt.axes(aspect='equal')
plt.scatter(test_labels, test_predictions)
plt.xlabel('True values (expenses)')
plt.ylabel('Predictions (expenses)')
lims = [0, 50000]
plt.xlim(lims)
plt.ylim(lims)
_ = plt.plot(lims,lims)
```
